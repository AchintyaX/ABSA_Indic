```
def get_polar_words(filename):
    with open(filename, encoding="ISO-8859-1") as f:
        words = set(f.read().splitlines())
    return words

positive_words = get_polar_words(PATH+"/positive-words.txt")
negative_words = get_polar_words(PATH+"/negative-words.txt")
```

```
pos_words = pos_words.rename(columns = {'SynsetTerms':'Term'})
neg_words = neg_words.rename(columns = {'SynsetTerms':'Term'})
```

```
positive_words = list(positive_words)
type(positive_words)
```

```
num_words = 0
for i in pos_words['Term']:
    if i not in positive_words: 
        num_words = num_words +1
num_words
```

```
pos = pos_words.to_dict('record')
pos
```

```
pos_array = []
for i in pos:
    dictionary = {}
    if i['Term'] not in positive_words:
        dictionary['Term'] = i['Term']
        dictionary['PosScore'] = i['PosScore']
        dictionary['NegScore'] = i['NegScore']
        pos_array.append(dictionary)
len(pos_array)
```

```
neg = neg_words.to_dict('record')
neg
```

```
positive_words = list(positive_words)
array = []
for i in positive_words:
    dictionary = {}
    dictionary['Term'] = i
    dictionary['PosScore'] = 0.90
    dictionary['NegScore'] = 0.00
    array.append(dictionary)
len(array)
```

```
pos_array.extend(array)
len(pos_array)
```

```
pos_df = pd.DataFrame(pos_array)
pos_df
```

```
neg_array = []
for i in neg:
    dictionary = {}
    if i['Term'] not in negative_words:
        dictionary['Term'] = i['Term']
        dictionary['PosScore'] = i['PosScore']
        dictionary['NegScore'] = i['NegScore']
        neg_array.append(dictionary)
len(neg_array)
```

```
array = []
for i in negative_words:
    dictionary = {}
    dictionary['Term'] = i
    dictionary['PosScore'] = 0.00
    dictionary['NegScore'] = -0.90
    array.append(dictionary)
len(array)
```

```
neg_array.extend(array)
len(neg_array)
```

```
neg_df = pd.DataFrame(neg_array)
neg_df
```

```
pos_df['hindi'] = [translator(i, 'hi') for i in pos_df['Term']]
pos_df
```

```
neg_df['hindi'] = [translator(i, 'hi') for i in neg_df['Term']]
neg_df
```

```
pos_df.to_csv(PATH+'/pos_words_hi.csv', index=False)
neg_df.to_csv(PATH+'/neg_words_hi.csv', index=False)
```
